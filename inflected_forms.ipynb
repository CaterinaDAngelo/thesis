```python

def pos_tag_dict_lemma(dict_lemma):
    noun_pattern = r'\(?fem|masc|neut\b'
    verb_pattern = r'\(?pres|imperf|aor|perf|fut|futperf|plup\b'
    for item in dict_lemma.items():
        if re.match(noun_pattern, item[1][0]):
            return "n"
        elif re.match(verb_pattern, item[1][0] ):
            return "v"
```

```python
noun_dict = {"lemma": [],
             "form": [],
             "gender": [],
             "case": [],
             "number": [],
             "dialects": []}

def get_noun_dict(lemma_data, lemma, noun_dict):
    # noun_dict["lemma"].extend([lemma] * len(lemma_data))
    pat_noun =  r'\(?fem|masc|neut'
    pat_numb = r'sg|pl|dual'
    for form, morphs in lemma_data.items():
        for morph in morphs:
            morph_list = morph.strip("()").split(" ")
            if len(morph_list)>=3:
                if re.match(pat_noun, morph_list[0]) and re.match(pat_numb, morph_list[2]):
                    noun_dict["lemma"].append(lemma)
                    noun_dict["form"].append(form)
                    noun_dict["gender"].append(morph_list[0])
                    noun_dict["case"].append(morph_list[1])
                    noun_dict["number"].append(morph_list[2])
                    if len(morph_list) > 3:
                        noun_dict["dialects"].append(morph_list[3:])
                    else:
                        noun_dict["dialects"].append(" ")

        
```

```python
tense_pattern = r'\(?pres|imperf|aor|perf|fut|futperf|plup'
mode_pattern = r"\(?inf|part|ind|subj|opt|imperat"
act_pattern = r"\(?act\b|mid\b|mp\b"
gender_pattern = r"\(?fem|masc|neut"
case_pattern = r"\(?nom|gen|dat|acc|voc"
person_pattern = r"\(?1st|2nd|3rd"
number_pattern = r"\(?sg|pl\b|dual"
dialects_pattern = r"\(?dor|aeo|hom|poe|ion|pros|parad"
```

```python
def find_match(pattern, string_list):
    """
    Finds the first string in the list that matches the given regex pattern.

    :param pattern: Regex pattern to search for.
    :param string_list: List of strings to search through.
    :return: First matching string or None if no match is found.
    """
    compiled_pattern = re.compile(pattern)
    for string in string_list:
        if compiled_pattern.search(string):
            return string
    return " "
```

```python
def get_verb_dict(lemma_data, lemma, verb_dict):
    for form, morphs in lemma_data.items():
        for morph in morphs:
            try:
                verb_dict["lemma"].append(lemma)
                verb_dict["form"].append(form)
                morph_list = morph.strip("()").split(" ")
                verb_dict["tense"].append(find_match(tense_pattern, morph_list))
                verb_dict["mode"].append(find_match(mode_pattern, morph_list))
                verb_dict["act/mid/p"].append(find_match(act_pattern, morph_list))
                verb_dict["gender"].append(find_match(gender_pattern, morph_list))
                verb_dict["case"].append(find_match(case_pattern, morph_list))
                verb_dict["person"].append(find_match(person_pattern, morph_list))
                verb_dict["number"].append(find_match(number_pattern, morph_list))
                if find_match(dialects_pattern, morph_list) != " ":
                    dialects_id = morph_list.index(find_match(dialects_pattern, morph_list))
                    verb_dict["dialects"].append(morph_list[dialects_id:])
                else:
                    verb_dict["dialects"].append(" ")
            except:
                print(f"problema con il lemma {morph_list}")
```
