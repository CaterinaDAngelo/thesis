```python
import cltk
```

```python
nlp = cltk.NLP(language="grc")
```

```python
docs = []
for file in file_directories: 
    if file.endswith(".txt"):
        lemmata_doc = []
        for line in codecs.open(f'./corpus/{file}', 'r+', encoding="utf-8"):
            line_anal = nlp.analyze(line)
            for lemma in line_anal.tokens:
                if lemma not in stop_words and "\\" not in lemma:
                    lemmata_doc.append(lemma)
        doc_no_stop = " ".join(lemmata_doc)
        docs.append(doc_no_stop)
```

```python
docs
```

```python
tokens = []
for doc in docs:
    for token in doc:
        tokens.append(token)
```

```python
len(tokens)
```

```python
#compute bigrams
from gensim.models import Phrases
phrases = Phrases(docs, min_count = 10, threshold=2)
```

```python
# Get the vocabulary from the Phrases model
vocab = phrases.vocab

for idx in range(len(docs)):
    for token in phrases[docs[idx]]:
        if '_' in token:
            # Token is a bigram, check the score
            score = vocab.get(token)
            if score > 10:
                docs[idx].append(token)
```

```python
# check the bigrams
bigrams = []
for phrase, score in vocab.items():
    if "_" in phrase:
        if score > 10:
            bigrams.append(phrase)

bigrams
```

```python
from gensim.corpora import Dictionary
```

```python
# Create a dictionary representation of the documents.
dictionary = Dictionary(docs)

print("Dictionary lenght 1):", len(dictionary))

# Filter out words that occur less than 5 documents, or more than 50% of the documents.
dictionary.filter_extremes(no_below=5, no_above=0.5)


print("Dictionary lenght 2):",len(dictionary))
```

```python
docs = [[token for token in doc if not token.isnumeric()] for doc in docs]
```

```python
# Bag-of-words representation of the documents.
corpus = [dictionary.doc2bow(doc) for doc in docs]
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

```python
from gensim.models import HdpModel
```

```python
# We first try with Hierarchical Dirichlet Process to see what would be the optimal number of topics
temp = dictionary[0] 
id2word = dictionary.id2token

hdp = HdpModel(
    corpus=corpus,
    id2word=id2word
)
```

```python
from gensim.models import LdaModel
```

```python
num_topics = 15
chunksize = 2000
passes = 20
iterations = 200

# Make an index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token

lda_model = LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=100,
    random_state=4
)

top_topics = lda_model.top_topics(corpus, topn=20)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

for topic in top_topics:
    print()
    print (topic[1], [w[1] for w in topic[0]])
```

```python
cltk_nlp = cltk.NLP(language="grc")
```

```python
def lemmatize_cltk(string):
    doc = cltk_nlp.analyze(text=string)
    lemmas = [lemma for lemma in doc.lemmata]
    return lemmas
```

```python
docs2predict = []
for string in strings:
    temp_lemmata = lemmatize_cltk(string)
    lemmata = []
    for lemma in temp_lemmata:
        if lemma not in stop:
            lemmata.append(lemma)
    docs2predict.append(lemmata)
            
            
```

```python
def predict_topics(sentence, topic_words):
    lemmata = set([lemma for lemma in lemmatize_cltk(sentence) if lemma not in stop])
    overlaps = {
        topic_id: len(lemmata & words)
        for topic_id, words in topic_words.items()
    }
    predicted_topic = max(overlaps, key=overlaps.get)
    return predicted_topic, overlaps[predicted_topic]


    
```
